```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
df[['COVID-19 Deaths', 'Number of Mentions']].plot(kind='bar')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


condition_deaths = df.groupby('Condition Group')['COVID-19 Deaths'].sum()

# Plot a bar plot for 'condition' vs 'no of deaths'
plt.figure(figsize=(10, 6))
condition_deaths.plot(kind='bar')
plt.xlabel('Condition')
plt.ylabel('Total Number of Deaths')
plt.title('Total Number of Deaths by Condition')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```

```python
import seaborn as sns

heatmap_data = df.pivot_table(index='State', values='COVID-19 Deaths', aggfunc='sum')

plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, fmt='g', cmap='YlGnBu')
plt.title('Number of Deaths by State')
plt.xlabel('Number of Deaths')
plt.ylabel('State')
plt.show()
```

```python
age_deaths = df.groupby('Age Group')['COVID-19 Deaths'].sum()

# Plot a bar plot for 'condition' vs 'no of deaths'
plt.figure(figsize=(10, 6))
age_deaths.plot(kind='bar')
plt.xlabel('age')
plt.ylabel('Total Number of Deaths')
plt.title('Total Number of Deaths by age')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```

```python
age_cases = df.groupby('Age Group')['Number of Mentions'].sum()

# Plot a bar plot for 'condition' vs 'no of cases'
plt.figure(figsize=(10, 6))
age_cases.plot(kind='bar')
plt.xlabel('age')
plt.ylabel('Total Number of cases')
plt.title('Total Number of cases by age')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```
